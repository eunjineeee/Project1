```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./datasets/df_final_0627.csv')
df
```

```python
df.drop('금값종가', axis = 1, inplace= True)
```

```python
gold = pd.read_csv('./datasets/XAU_USD 과거 데이터.csv')
gold
```

```python
# 필요한 컬럼만 추출
gold = gold[['날짜', '종가']]
gold
```

```python
# 컬럼이름 변경
gold = gold.rename(columns={'날짜' : '계약일', '종가' : '금값종가'})
```

```python
# 시계열 변경
df['계약일'] = pd.to_datetime(df['계약일'])
gold['계약일'] = pd.to_datetime(gold['계약일'])
```

```python
# 합치기
merged_df1 = pd.merge(df, gold, on='계약일', how='left')
```

```python
merged_df1
```

```python
# LTV 파생변수 생성
merged_df1['LTV'] = np.NaN

# 타입 변환
merged_df1['계약일'] = pd.to_datetime(merged_df1['계약일'])

merged_df1.loc[merged_df1['계약일'] <= '2017-06-19', 'LTV'] = 0.7

# 17.06.19 
merged_df1.loc[(merged_df1['계약일'] > '2017-06-19')& (merged_df1['계약일'] <= '2017-08-02'), 'LTV'] = 0.6

# 17.08.02
merged_df1.loc[(merged_df1['계약일'] > '2017-08-02')&(merged_df1['계약일'] <= '2018-09-14'), 'LTV'] = 0.4

# 18.09.13 -> 1주택 가정 
    # 서울 전체 주택보유자 비율 훨씬 높음 2:1 
    # 부동산 거래Data 중에서도 서울 APT를 사는 사람은 기존 보유주택을 팔고 온것이라 가정
# merged_df1.loc[(merged_df1['계약일'] > '2018-09-14')& (merged_df1['계약일'] <= '2019-12-15'), 'LTV'] = 0.4
# merged_df1.loc[((merged_df1['계약일'] > '2018-09-14')& (merged_df1['계약일'] <= '2019-12-15')) & (merged_df1['물건금액(만원)']>90000), 'LTV'] = 0
merged_df1.loc[(merged_df1['계약일'] > '2018-09-14')& (merged_df1['계약일'] <= '2019-12-16'), 'LTV'] = 0

# 19.12.16
merged_df1.loc[(merged_df1['계약일'] > '2019-12-16'), 'LTV'] = 0.4
merged_df1.loc[(merged_df1['계약일'] > '2019-12-16') & (merged_df1['물건금액(만원)'] > 90000), 'LTV'] = 0.2
merged_df1.loc[(merged_df1['계약일'] > '2019-12-16') & (merged_df1['물건금액(만원)'] > 150000), 'LTV'] = 0
```

```python
# '물건금액(만원)' 컬럼을 마지막으로 이동시키기
cols = merged_df1.columns.tolist()
cols.remove('물건금액(만원)')
cols.append('물건금액(만원)')
merged_df1 = merged_df1[cols]
merged_df1
```

```python
merged_df1['금값종가'] = merged_df1['금값종가'].fillna(method='ffill')
merged_df1
```

```python
merged_df1.to_csv('df_all_0629.csv', index = False, encoding='utf-8-sig')
```
